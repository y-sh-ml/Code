### 3.

```python
# カラム内の文字を数字に変換する

def code_transform(x):
    if x == 'F':
        y = 0
    elif x == 'M':
        y = 1
    else:
        y = 2
        
    return y

df['CODE_GENDER_INT'] = df.CODE_GENDER.apply(lambda x: code_transform(x))
df['CODE_GENDER_INT'].head()
```

```python
# \ で改行できる

# AMT は平均、REGION は最小値を求める

df[['AMT_INCOME_TOTAL', 'REGION_POPULATION_RELATIVE']] \
.apply(lambda x: np.mean(x) if x.name == 'AMT_INCOME_TOTAL' else np.min(x), axis=0)
```

### 6.

```python
# 数値データと文字データのカラムを分ける（前処理で頻繁に使用する）

numerical_col = []
categorical_col = []

for col_name in df.columns:
    if df[col_name].dtype == 'object':
        categorical_col.append(col_name)
    else:
        numerical_col.append(col_name)
```

```python
#リスト内包表記でも可能

numerical_col = [col for col in df.columns if df[col].dtypes != 'object']
categorical_col = [col for col in df.columns if df[col].dtypes == 'object']
```

```python
# リスト型は shape が使えないので、len() を使用する

print(len(numerical_col))
print(len(categorical_col))
```

```python
# 使用するカラムを決め、データフレームを整える

bureau_group = bureau[['SK_ID_CURR', 'SK_ID_BUREAU', 'DAYS_CREDIT']] \
    .query('SK_ID_CURR == 215354 | SK_ID_CURR == 162297') \
    .groupby(by = 'SK_ID_CURR') \
    .apply(lambda x: x.sort_values(['DAYS_CREDIT'], ascending = False)) \
    .reset_index(drop=True)
```

```python
bureau_group.head(10)
```

```python
bureau_group['DIFF'] = bureau_group.groupby(by = ['SK_ID_CURR'])['DAYS_CREDIT'].shift(-1)

bureau_group['PREVIOUS_DAYS_DIFF'] = np.abs(bureau_group['DAYS_CREDIT'] - bureau_group['DIFF'])

bureau_group.fillna(0, inplace=True)

bureau_group
```
