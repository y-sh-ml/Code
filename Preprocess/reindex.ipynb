```python
import pandas as pd
```

```python
df = pd.DataFrame({'A': [1, 2, 3], 'B': [10, 20, 30], 'C': [100, 200, 300]}, index=['One', 'Two', 'Three'])
df
```

```python
df.reindex(index=['Two', 'Three', 'One'])
```

```python
df.reindex(columns=['B', 'C', 'A'])
```

```python
df.reindex(index=['Two', 'Three', 'One'], columns=['B', 'C', 'A'])
```
